# Análisis del Diagrama de Dispersión: Temperatura vs. Bicicletas Rentadas

Este diagrama de dispersión muestra la relación entre la **temperatura (°C)** y el **número de bicicletas rentadas**.

## Interpretación:
- **Relación positiva**: A medida que la temperatura aumenta, también lo hace el número de bicicletas rentadas. Esto indica que existe una correlación positiva entre ambas variables.
- **Puntos concentrados**: La mayor concentración de bicicletas rentadas ocurre entre los 15°C y 25°C, lo que sugiere que las personas prefieren alquilar bicicletas cuando el clima es más cálido.
- **Valores bajos de temperatura**: A temperaturas por debajo de 0°C, el número de bicicletas rentadas disminuye drásticamente, lo que podría indicar que el frío extremo desalienta el alquiler.
- **Valores extremos**: En temperaturas muy altas (cerca de los 30°C), aunque todavía hay muchas bicicletas rentadas, se observa una ligera dispersión, lo que podría sugerir que las condiciones extremas de calor también afectan el número de alquileres.


```python
# Diagramas de violín: Distribución de bicicletas rentadas por estación, con paleta de colores
plt.figure(figsize=(10, 6))
sns.violinplot(x='Seasons', y='Rented Bike Count', data=df, palette="Set2")
plt.title('Diagrama de Violín: Bicicletas Rentadas por Estación', fontsize=16)
plt.show()
```

# Análisis del Diagrama de Violín: Bicicletas Rentadas por Estación

Este diagrama de violín muestra la distribución del **número de bicicletas rentadas** a lo largo de las diferentes **estaciones del año**.

## Interpretación:
- **Winter (Invierno)**: Tiene la menor cantidad de bicicletas rentadas, con una distribución muy compacta y concentrada entre 0 y 500 bicicletas. Esto sugiere que el invierno no es una temporada popular para alquilar bicicletas.
- **Spring (Primavera)**: Presenta una mayor variabilidad y un aumento considerable en el número de bicicletas rentadas, con un rango que va desde 0 hasta 3500 bicicletas. La primavera parece ser una de las estaciones más populares para el alquiler de bicicletas.
- **Summer (Verano)**: Similar a la primavera, el verano muestra una gran cantidad de bicicletas rentadas, con una distribución ligeramente más amplia. Sin embargo, los alquileres tienden a concentrarse más cerca del rango de 1000-2000 bicicletas.
- **Autumn (Otoño)**: Aunque no es tan popular como la primavera y el verano, el otoño muestra un patrón similar, con una distribución más baja que en las estaciones cálidas, pero aún con una amplia variabilidad.

## Conclusión:
- Las estaciones cálidas, **primavera** y **verano**, tienen la mayor cantidad de bicicletas rentadas, con una distribución más amplia y una gran variabilidad.
- **Invierno** muestra la menor actividad de alquiler de bicicletas, lo que indica que las condiciones invernales no son favorables para este tipo de transporte.


```python
import matplotlib.pyplot as plt

# Crear un gráfico de dispersión para cada variable numérica contra el número de bicicletas rentadas
variables = ['Hour', 'Temperature(C)', 'Humidity(%)', 'Wind speed (m/s)', 'Visibility (10m)', 
             'Dew point temperature(C)', 'Solar Radiation (MJ/m2)', 'Rainfall(mm)', 'Snowfall (cm)']

plt.figure(figsize=(15, 20))

for i, variable in enumerate(variables, 1):
    plt.subplot(5, 2, i)
    sns.scatterplot(x=df[variable], y=df['Rented Bike Count'], alpha=0.6, color='green')
    plt.title(f'Relación entre {variable} y # Bicicletas Rentadas')
    plt.xlabel(variable)
    plt.ylabel('Bicicletas Rentadas')

plt.tight_layout()
plt.show()
```

# Análisis de la Relación entre Variables y el Número de Bicicletas Rentadas

Este análisis breve muestra la relación entre el número de bicicletas rentadas y cada una de las variables presentes en el conjunto de datos.

## 1. **Hour (Hora) vs. Rented Bike Count (Bicicletas Rentadas)**
   - **Interpretación**: El número de bicicletas rentadas aumenta significativamente en horas específicas del día, especialmente entre las 7-9 AM y de nuevo entre las 5-8 PM. Esto probablemente refleja los patrones de desplazamiento diario (horas pico), ya que muchas personas podrían usar las bicicletas para ir y volver del trabajo.

## 2. **Temperature(C) (Temperatura en grados Celsius) vs. Rented Bike Count (Bicicletas Rentadas)**
   - **Interpretación**: A medida que la temperatura aumenta, el número de bicicletas rentadas también tiende a aumentar, con una mayor concentración de bicicletas rentadas entre los 15°C y 25°C. Las temperaturas más bajas parecen desalentar el uso de bicicletas, lo cual es esperable dado que las condiciones climáticas frías no son ideales para el ciclismo.

## 3. **Humidity(%) (Humedad) vs. Rented Bike Count (Bicicletas Rentadas)**
   - **Interpretación**: La humedad no parece tener una relación clara con el número de bicicletas rentadas. El número de bicicletas rentadas se mantiene distribuido a lo largo de todos los niveles de humedad, lo que sugiere que este factor no influye fuertemente en el comportamiento de alquiler.

## 4. **Wind speed (m/s) (Velocidad del viento) vs. Rented Bike Count (Bicicletas Rentadas)**
   - **Interpretación**: Aunque la mayoría de los registros se concentran en velocidades de viento bajas (0-2 m/s), no se observa una relación clara entre la velocidad del viento y el número de bicicletas rentadas. Sin embargo, podría haber menos bicicletas rentadas en días con vientos más fuertes, lo que se refleja en la dispersión de puntos más alta con mayor velocidad del viento.

## 5. **Visibility (10m) (Visibilidad) vs. Rented Bike Count (Bicicletas Rentadas)**
   - **Interpretación**: La relación entre la visibilidad y el número de bicicletas rentadas parece positiva. A mayor visibilidad, más bicicletas son rentadas, lo cual es lógico, ya que las condiciones de baja visibilidad pueden disuadir a los usuarios de alquilar bicicletas.

## 6. **Dew point temperature(C) (Temperatura del punto de rocío) vs. Rented Bike Count (Bicicletas Rentadas)**
   - **Interpretación**: A medida que la temperatura del punto de rocío aumenta, el número de bicicletas rentadas también parece aumentar. Esto podría estar relacionado con un clima más cálido y agradable cuando el punto de rocío es más alto, lo que favorece el uso de bicicletas.

## 7. **Solar Radiation (MJ/m²) (Radiación solar) vs. Rented Bike Count (Bicicletas Rentadas)**
   - **Interpretación**: Aunque no hay una tendencia muy fuerte, parece que los días con mayor radiación solar tienden a tener más bicicletas rentadas. Esto sugiere que los días soleados son más propicios para alquilar bicicletas.

## 8. **Rainfall (mm) (Precipitación) vs. Rented Bike Count (Bicicletas Rentadas)**
   - **Interpretación**: Hay una clara tendencia inversa: a mayor cantidad de lluvia, el número de bicicletas rentadas disminuye drásticamente. La mayoría de los días con precipitaciones muestran pocas bicicletas rentadas, lo cual es natural, ya que la lluvia es una barrera significativa para el ciclismo.

## 9. **Snowfall (cm) (Nevadas) vs. Rented Bike Count (Bicicletas Rentadas)**
   - **Interpretación**: Similar a la lluvia, las nevadas tienen un efecto negativo en el número de bicicletas rentadas. En días con nieve, se observan muy pocas bicicletas alquiladas, lo que sugiere que las condiciones invernales extremas no son favorables para el uso de bicicletas.

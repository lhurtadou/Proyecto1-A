```python
# Function to remove outliers based on the IQR method
def remove_outliers(df, columns):
    for column in columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
    
        df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df

```

```python
# Columns to check for outliers
columns_to_check = [
    'Rented Bike Count', 
    'Temperature(C)', 
    'Humidity(%)', 
    'Wind speed (m/s)', 
    'Visibility (10m)', 
    'Dew point temperature(C)', 
    'Solar Radiation (MJ/m2)', 
    'Rainfall(mm)', 
    'Snowfall (cm)'
]
```

```python

# Remove outliers from the dataset
cleaned_data = remove_outliers(df, columns_to_check)
```

```python
X= cleaned_data.drop(columns='Rented Bike Count')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn import metrics
import numpy as np

# Separar variables predictoras (X) y la variable objetivo (y)
y = cleaned_data['Rented Bike Count']

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.24, random_state=1)

# Crear y entrenar el modelo
model = LinearRegression()
model.fit(X_train, y_train)

# Realizar predicciones
y_pred = model.predict(X_test)
```

```python
import numpy as np

# mean absolute error
MAE = metrics.mean_absolute_error(y_test, y_pred)

# mean squared error
MSE = metrics.mean_squared_error(y_test, y_pred)

# root mean squared error
RMSE = np.sqrt(MSE)
```

```python
print("MAE: ", MAE)
print("MSE: ", MSE)
print("RMSE: ", RMSE)
```
